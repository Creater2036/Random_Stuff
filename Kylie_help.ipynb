```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import plotly.express as px
```

```python
df = pd.read_csv('Comprehensive_Data_List_Cleaned_AD.csv')
df.head()
```

```python
plt.figure(figsize = (12, 7), dpi = 150);
sns.countplot(data = df.nlargest(30, 'Overall Rating'), x = 'Cause')
plt.xticks(rotation = 60);
plt.title('Number of Causes for the Highest 30 Rated Charities')
```

```python
plt.figure(figsize = (12, 7), dpi = 150);
sns.countplot(data = df.nlargest(30, 'Overall Rating'), x = 'State')
plt.xticks(rotation = 60);
plt.title('Location of Charity for the Highest 30 Rated Charities')
```

```python
plt.figure(figsize = (14, 11), dpi = 300);
sns.heatmap(df.pivot_table(index = 'Cause', columns = 'State', values='Overall Rating', aggfunc = 'count', fill_value=0), robust = True, annot = True, fmt = 'g', annot_kws={"fontsize":8})
plt.title('Heatmap of Number of Charities based on State and Cause');
```
